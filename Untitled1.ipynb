```python
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de ejemplo para y_true (etiquetas verdaderas) y y_pred (etiquetas predichas)
# Asumimos un problema de clasificación con 3 clases (0, 1, 2)
np.random.seed(42) # para reproducibilidad
y_true = np.random.randint(0, 3, size=100) # 100 muestras, 3 clases
y_pred = np.random.randint(0, 3, size=100) # 100 muestras, 3 clases

print("Primeros 10 valores de y_true:", y_true[:10])
print("Primeros 10 valores de y_pred:", y_pred[:10])
```

```python
# Calcular la matriz de confusión
mc = confusion_matrix(y_true, y_pred)

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(mc, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### Revisión de Valores Nulos y Tipos de Datos

Vamos a revisar el estado actual de los valores nulos y los tipos de datos de cada columna para planificar las siguientes acciones de limpieza.

```python
print("Valores faltantes por columna (actualizado):")
print(df.isnull().sum())

print("\nTipos de datos de las columnas:")
print(df.info())
```

```python
# Calcular el porcentaje de valores nulos para cada columna
missing_percentage = df.isnull().sum() / len(df) * 100

# Identificar columnas con más del 90% de valores nulos
columns_to_drop = missing_percentage[missing_percentage > 90].index.tolist()

if columns_to_drop:
    print(f"Columnas a eliminar por tener más del 90% de valores nulos: {columns_to_drop}")
    df = df.drop(columns=columns_to_drop)
    print("Columnas eliminadas con éxito.")
else:
    print("No se encontraron columnas con más del 90% de valores nulos para eliminar.")

# Mostrar las columnas restantes
print("\nColumnas restantes:")
print(df.columns.tolist())
```
